```python
Q1=df['Age'].quantile(0.25)
print(Q1)
Q3=df['Age'].quantile(0.75)
print(Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print(lower_bound)
print(upper_bound)
```

```python
df_out=df_new[(df_new['Age']>lower_bound) & (df_new['Age']<upper_bound)]
```

```python
sns.boxplot(data=df_out,x='Age')
```

```python
sns.boxplot(data=df_out)
```
