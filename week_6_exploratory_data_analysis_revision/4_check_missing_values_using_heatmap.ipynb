### seaborn heatmap
* several methods to deal with missing values
* heatmap
* boxplot: to check outliers and missing values
* scatter plot: 
* density plots:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic_train.csv")
df.head()
```

```python
sns.heatmap(df.isnull())
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
sns.histplot(df)
```

```python
sns.boxplot(data=df)
```

```python
sns.scatterplot(data=df)
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), yticklabels=False,cbar=False,cmap='viridis')
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df,dodge=False)
```

```python
# this box plot is showing outliers.(here fare class has the highest no of outliers)
```

```python
sns.displot(data=df)
```
