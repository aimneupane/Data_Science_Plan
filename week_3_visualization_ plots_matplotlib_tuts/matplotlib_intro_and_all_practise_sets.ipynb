### Matplot library use

```python
# it is a low level graph plotting library that serves as visualization utility.
# open source and free.
# created by john D. hunter
```

```python
# most of the matplotlib utilities lies under the pyplot submodule.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# let us take an example.
x=np.array([0,6])
y=np.array([0,250])
plt.plot(x,y)
plt.show()
```

```python
# to plot only the markers , we can use shortcut notation 'o'
plt.plot(x,y,'o')
plt.show()
```

```python
# we can provide the size and color of marker too.
plt.plot(x,y,marker='o',ms=20,mec='r',mfc='r')
plt.show()
#here mec means marker edge color
#mfc means marker face color.
```

```python
# we can use line style to change the style of plotted line.
plt.plot(x,y,linestyle='dotted')
plt.show()
```

```python
# we can use labels and title  too
plt.plot(x,y)
plt.title("calorie burn average")
plt.xlabel("average pulse")
plt.ylabel("calorie burns")
plt.grid()
plt.show()
```
