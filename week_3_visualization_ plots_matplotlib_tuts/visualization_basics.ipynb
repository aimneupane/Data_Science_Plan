```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot("survived",data=df)
```

```python
sns.countplot("survived",hue="sex",data=df)
```

```python
#scatter plot
plt.scatter(data=df,x="age",y="fare")
```

```python
sns.scatterplot(data=df,x="age",y="fare",hue="pclass")
```

```python
sns.barplot(data=df,x="sex",y="fare",hue="pclass")
```

```python
sns.pairplot(df)
```

```python
df.hist(bins=10,figsize=(9,9),grid=False)
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(data=df,x="sex",y="fare",hue="pclass")
```

```python
sns.boxenplot(data=df,x="sex",y="fare",hue="pclass")
```

```python
g=sns.FacetGrid(df,hue="survived",col="pclass",margin_titles=True,palette={1:"seagreen",0:"gray"})
g=g.map(plt.scatter,"fare","age",edgecolor="w").add_legend()
```

```python
#heat map
corr=df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
```
