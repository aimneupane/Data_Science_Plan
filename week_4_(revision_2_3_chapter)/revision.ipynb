### check null values tomorrow

```python
df.isna()
#here in data if false is written then data is okay 
#but if true is shown then there must be null values
# this isna() shows all every data in a dataframe
```

```python
df.isna().any()
#it comprises in only column data that if there is null or not null
```

```python
df.isna().sum()
#here sum tells it adds and shows how many data have null values
```
